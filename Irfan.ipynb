```python
data1=data[['course_id','price','num_subscribers','num_reviews','num_lectures','level','content_duration','published_timestamp','subject']]
print(type(data1))
data1
```

```python
df=pd.DataFrame(data1)
df.head()
```

```python
sns.pairplot(df,hue="level")
plt.show()
```

```python
sns.boxplot(x="price" , y='subject' , data=df) 
plt.show()
```

```python
final=data.pivot_table(index = ["level","subject"], columns = ["price"], values = "num_subscribers" )
```

```python
sns.heatmap(final, cmap="coolwarm", )
plt.show()
```

```python
beg=data["level"]=="Beginner Level"
beg1=data["subject"]=="Business Finance"
result=data[beg & beg1]
sns.jointplot(x=result['price'], y=result['num_subscribers'], data =result , kind='scatter')
plt.show()
beg=data["level"]=="All Levels"
beg1=data["subject"]=="Graphic Design"
result=data[beg & beg1]
sns.jointplot(x=result['price'], y=result['num_subscribers'], data =result , kind='scatter')
plt.show()
beg=data["level"]=="Expert Level"
beg1=data["subject"]=="Graphic Design"
result=data[beg & beg1]
sns.jointplot(x=result['price'], y=result['num_subscribers'], data =result , kind='scatter')
plt.show()
beg=data["level"]=="Intermediate Level" 
beg1=data["subject"]=="Graphic Design"
result=data[beg & beg1]
sns.jointplot(x=result['price'], y=result['num_subscribers'], data =result , kind='scatter')
plt.show()
```
